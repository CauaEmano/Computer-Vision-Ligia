## Importação das bibliotecas usadas

```python
import tensorflow as tf
import keras_tuner as kt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow import keras
```

## Pré-Processamento e Data Augmentation

```python
#Evitei alterações bruscas no datagen de treino por ser um teste de raio-x, permitir horizontal-flip/vertical-flip poderia 
#acarretar na construção de correlações errôneas, então priorizei pequenas mudanças para apenas "dificultar" a vida do modelo durante o treino
datagen_treino = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 20,
    zoom_range = 0.1,
    horizontal_flip = False,
    vertical_flip = False,
    validation_split = 0.2 #Parte separada para a validação
)

datagen_teste = ImageDataGenerator(rescale=1./255)
```

```python
#Extraindo as imagens diretamente das pastas
train_generator = datagen_treino.flow_from_directory(
    directory = 'train/train',
    target_size = (256, 256),
    batch_size = 16, #Separei em batch de tamanho 16 em prol da minha gpu que não estava conseguindo fazer o tuning certinho com 32 ks
    class_mode = 'categorical',
    shuffle = True, 
    subset = 'training',
    seed = 42
)

val_generator = datagen_treino.flow_from_directory(
    directory = 'train/train',
    subset = 'validation',
    shuffle = False, 
    seed = 42 #Sem embaralhar para garantir a reprodutibilidade
)    

teste_generator = datagen_teste.flow_from_directory(
    directory = 'test_images', 
    target_size = (256,256),
    batch_size = 16,
    class_mode = 'categorical', 
    shuffle = False
)
```

## Otimização de Hiperparâmetros usando KerasTuner

```python
def construir_modelo(hparams):
    modelo = keras.Sequential() #Instância do modelo

    #Construir as noções mais básicas dos dados
    filtros_1 = hparams.Int('filtros_1', min_value=32, max_value=96, step=32) #Quantidade de filtros a serem utilizados na convolução incial
    modelo.add(keras.layers.Conv2D(filtros_1, (3,3), input_shape=(256,256,3), activation='relu'))
    modelo.add(keras.layers.MaxPooling2D())

    #Construir as noções mais complexas dos dados
    filtros_2 = hparams.Int('filtros_2', min_value=64, max_value=128, step=32) #Quantidade de filtros a serem utilizados na convolução seguinte
    modelo.add(keras.layers.Conv2D(filtros_2, (3,3), activation='relu'))
    modelo.add(keras.layers.MaxPooling2D())

    filtros_3 = hparams.Int('filtros_3', min_value=64, max_value=128, step=32) 
    modelo.add(keras.layers.Conv2D(filtros_3, (3,3), activation='relu'))
    modelo.add(keras.layers.MaxPooling2D())

    #Redimensionando os dados para serem interpretados
    modelo.add(keras.layers.Flatten())

    #Testando diferentes níveis de dropout, a fim de evitar overfitting
    taxa_dropout = hparams.Float('dropout', min_value=0.0, max_value=0.5, step=0.1)
    modelo.add(keras.layers.Dropout(taxa_dropout))

    #Camada final de output
    modelo.add(keras.layers.Dense(2, activation='softmax'))

    lr = hparams.Choice('learning_rate', values=[0.01, 0.001, 0.0001])

    modelo.compile( #Compilação
        optimizer = keras.optimizers.Adam(learning_rate=lr),
        loss = 'categorical_crossentropy', 
        metrics = ['accuracy', 'AUC']
    )

    return modelo
```

```python
tuner = kt.RandomSearch( #Instanciando o tuner a ser utilizado
    construir_modelo,
    objective = kt.Objective('val_AUC', direction='max'), #Buscar o maior valor no ROC_AUC
    max_trials = 20,
    executions_per_trial = 1,
    directory = 'projeto_ligia_tuner',
    project_name = 'busca_pneumonia'
)

tuner.search_space_summary() 
```

```python
peso_classes = {0: 3.0, 1:1.0} #Atribuindo pesos nas penalidades entre as classes normal e pneumonia (evidenciado na EDA

historico = tuner.search(train_generator, epochs=5, validation_data=val_generator, class_weight=peso_classes)
```

```python
melhores_parametros = tuner.get_best_hyperparameters(num_trials=1)[0] #Extraindo os melhores hiperparâmetros
print(melhores_parametros.values)
```

## Construção do modelo final e geração do arquivo csv

```python
modelo_final = keras.Sequential() #Construção do modelo final, retreinando com mais epochs, arquitetura verificada pós otimização com KerasTuner
modelo_final.add(keras.layers.Conv2D(96, (3,3), input_shape=(256,256,3), activation='relu'))
modelo_final.add(keras.layers.MaxPooling2D())
modelo_final.add(keras.layers.Conv2D(128, (3,3), activation='relu'))
modelo_final.add(keras.layers.MaxPooling2D())
modelo_final.add(keras.layers.Conv2D(128, (3,3), activation='relu'))
modelo_final.add(keras.layers.MaxPooling2D())
modelo_final.add(keras.layers.Flatten())
modelo_final.add(keras.layers.Dropout(0.3))
modelo_final.add(keras.layers.Dense(2, activation='softmax'))

modelo_final.compile(
        optimizer = keras.optimizers.Adam(learning_rate=0.001),
        loss = 'categorical_crossentropy', 
        metrics = ['accuracy', 'AUC']
    )
```

```python
checkpoint = ModelCheckpoint ( #Salvar o peso do modelo de melhor desempenho baseado no ROC-AUC
    filepath = 'melhor_modelo.h5',
    monitor = 'val_AUC',
    mode = 'max',
    save_best_only = True,
    verbose = 1
)

reduce_lr = ReduceLROnPlateau( #Redução na taxa de aprendizagem em caso de estagnar
    monitor = 'val_AUC', 
    factor = 0.2,
    patience = 3,
    min_lr = 0.00001,
    verbose = 1
)

early_stop = EarlyStopping( #Interrompe antecipadamente o treino caso sinais de overfit
    monitor = 'val_AUC',
    patience = 8,
    mode = 'max', 
    verbose = 1,
    restore_best_weights = True
)

historico_final = modelo_final.fit( #Treinando...
    train_generator,
    epochs=20,
    validation_data = val_generator,
    class_weight = peso_classes,
    callbacks = [checkpoint, reduce_lr, early_stop]
)
```

### Geração dos plots

```python
acc = historico_final.history['accuracy'] 
val_acc = historico_final.history['val_accuracy']
loss = historico_final.history['loss']
val_loss = historico_final.history['val_loss']
epocas = range(1, len(acc)+1)

plt.figure(figsize=(14,5))

#Plot da acurácia
plt.subplot(1,2,1)
plt.plot(epocas, acc, 'b', label='acurácia no treino')
plt.plot(epocas, val_acc, 'r', label='acurácia na validação')
plt.title('Evolução da acurácia')
plt.xlabel('épocas')
plt.ylabel('acurácia')
plt.grid(True)
plt.legend()

#Plot da loss function
plt.subplot(1,2,2)
plt.plot(epocas, loss, 'b', label='perda no treino')
plt.plot(epocas, val_loss, 'r', label='perda na validação')
plt.title('Evolução da perda')
plt.xlabel('épocas')
plt.ylabel('loss')
plt.grid(True)
plt.legend()

plt.show()
```

```python
auc = historico_final.history['AUC']
val_auc = historico_final.history['val_AUC']

#Plot do AUC
plt.plot(epocas, auc, 'b', label='ROC-AUC no treino')
plt.plot(epocas, val_auc, 'r', label='ROC-AUC na validação')
plt.title('Evolução do ROC-AUC')
plt.xlabel('épocas')
plt.ylabel('AUC')
plt.grid(True)
plt.legend()

plt.show()
```

### Criação do arquivo csv com os testes preditos certinho

```python
import cv2
import matplotlib.cm as cm

def gerar_gradcam(img_path, modelo, ultima_camada_conv_nome):
    #Carrega e prepara a imagem
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(256, 256))
    img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    img_tensor = tf.convert_to_tensor(img_array, dtype=tf.float32)

    #Corta o modelo Sequencial em duas partes!
    indice_camada = None
    for i, layer in enumerate(modelo.layers):
        if layer.name == ultima_camada_conv_nome:
            indice_camada = i
            break

    if indice_camada is None:
        print(f"Erro: Camada {ultima_camada_conv_nome} não encontrada!")
        return

    modelo_parte1 = tf.keras.Sequential(modelo.layers[:indice_camada + 1])
    modelo_parte2 = tf.keras.Sequential(modelo.layers[indice_camada + 1:])

    #Usa o GradientTape rastreando perfeitamente a conexão
    with tf.GradientTape() as tape:
        conv_outputs = modelo_parte1(img_tensor)
        tape.watch(conv_outputs) #Obriga o Keras a olhar para essa variável!
        preds = modelo_parte2(conv_outputs)
        pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    #Gradientes agora funcionam!
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    #Monta o Mapa de Calor
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = heatmap.numpy()

    #Sobrepõe na imagem
    img_original = cv2.imread(img_path)
    img_original = cv2.resize(img_original, (256, 256))
    
    heatmap_red = cv2.resize(heatmap, (256, 256))
    heatmap_red = np.uint8(255 * heatmap_red)
    heatmap_color = cv2.applyColorMap(heatmap_red, cv2.COLORMAP_JET)
    
    imagem_sobreposta = cv2.addWeighted(img_original, 0.6, heatmap_color, 0.4, 0)
    
    #Plota
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Raio-X Original")
    plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.title("Grad-CAM (Foco da IA)")
    plt.imshow(cv2.cvtColor(imagem_sobreposta, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
